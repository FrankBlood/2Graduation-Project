```python
import jieba
import graphlab
import jieba.posseg as pseg
```

```python
def cut(query):
    query_cut = ''
    temp = pseg.cut(query)
    for word, flag in temp:
        query_cut = query_cut + ' ' + word + ' ' + flag
    return query_cut
```

```python
data['query_cut'] = data['query'].apply(cut)
```

```python
def search_cut(query):
    query_search_cut = ''
    temp = jieba.cut_for_search(query)
    for word in temp:
        query_search_cut = query_search_cut + ' ' + word 
    return query_search_cut
```

```python
data['query_search_cut'] = data['query'].apply(search_cut)
```
